```python
x = 10
y = 10
next_x = 9
next_y = 10
if(y == next_y-1):
    print(0)
elif(x == next_x-1):
    print(1)
elif(y == next_y+1):
    print(2)
elif(x == next_x+1):
    print(3)
```

```python
directon = 3
x =10
y = 10
j = 11
i = 11
if(directon == 0):
    if((j == x) and (i >= y)):
        print(0)
elif(directon == 1):
    if((j >= x) and (i == y)):
        print(1)
elif(directon == 2):
    if((j == x) and (i <= y)):
        print(2)
elif(directon == 3):
    if((j <= x) and (i == y)):
        print(3)
```
